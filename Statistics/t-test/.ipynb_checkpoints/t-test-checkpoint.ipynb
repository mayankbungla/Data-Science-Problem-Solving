```python
import math
import numpy as np
from scipy.stats import t
```

Q1 A sample of 20 students scored: mean=72, s=8. The national average is 70.             │
│  At α=0.05, is this school's performance significantly different?                        │
│  Use a one-sample t-test. Compute t, df, the critical value, and state your conclusion. │

```python
n = 20
x = 72
s= 8
mu = 70
alpha = 0.05
```

```python
t_stat = (x- mu)/(s/ np.sqrt(n))
t_stat
```

```python
df= n-1
```

```python
t_crit = t.ppf(1 - alpha/2,19)
t_crit
```

t_stat < t_crit
- Fail to reject null hypotheses

```python
p = 2*t.sf(abs(t_stat), df=19)    
```
