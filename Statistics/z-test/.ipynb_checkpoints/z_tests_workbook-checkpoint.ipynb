## 📌 Question 1 — One-Sample Z-Test (Mean)

A factory claims that the mean diameter of its bolts is **10 mm**.

A sample of **50 bolts** is taken, and the following statistics are observed:

| Parameter | Value |
|-----------|-------|
| Sample mean (x̄) | 10.3 mm |
| Population std dev (σ) | 1.0 mm |
| Sample size (n) | 50 |
| Significance level (α) | 0.05 |

Perform a **two-tailed one-sample Z-test** to evaluate the factory's claim.
Also compute the **95% confidence interval** for the mean.

---

#### Step 1 — Hypothesis

- **H₀:** μ = 10 (factory's claim holds)
- **H₁:** μ ≠ 10 (two-tailed — we don't know which direction the mean might shift)

We're not assuming the bolt is too big or too small specifically, so two-tailed makes sense here.

#### Step 2 — Why Z-Test?

- Population σ is **known** → Z-test (if σ were unknown, we'd use t-test)
- n = 50, which is large enough for CLT to kick in
- We'll check both critical value and p-value methods to cross-verify

```python
# Q1 given values
xbar_q1  = 10.3   # sample mean (mm)
mu0_q1   = 10     # hypothesized population mean
sig_q1   = 1      # known population std dev
n_q1     = 50     # sample size
a_q1     = 0.05   # significance level
```

#### Step 3 — Test Statistic

$$z = \frac{\bar{x} - \mu_0}{\sigma / \sqrt{n}}$$

We're measuring how many standard errors away the sample mean is from the claimed population mean.

```python
z_q1 = (xbar_q1 - mu0_q1) / (sig_q1 / n_q1**0.5)
z_q1
```

#### Step 4 — Critical Value Method

For α = 0.05 (two-tailed), we split the rejection region equally on both tails:

- Each tail gets α/2 = 0.025
- We need z such that P(Z < z) = 0.975 → use `norm.ppf(0.975)`

If |z_stat| > z_critical → reject H₀

```python
# critical z at alpha=0.05, two-tailed
zcrit_q1 = norm.ppf(1 - a_q1/2)
zcrit_q1
```

**Decision:** |z| = 2.12 > z_critical = 1.96 → **Reject H₀**

#### Step 5 — p-value Method

$$p = 2 \times P(Z > |z|)$$

`norm.sf(x)` gives P(Z > x) directly — the survival function. Multiply by 2 for two-tailed.

```python
# p-value for two-tailed test
pval_q1 = 2 * norm.sf(abs(z_q1))
pval_q1
```

**Decision:** p-value ≈ 0.034 < α = 0.05 → **Reject H₀** ✓ (consistent with critical value method)

#### Step 6 — 95% Confidence Interval

$$CI = \bar{x} \pm z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}$$

If μ₀ falls outside this interval, that's another way to reject H₀.

```python
moe_q1   = zcrit_q1 * (sig_q1 / np.sqrt(n_q1))  # margin of error
lo_q1    = xbar_q1 - moe_q1
hi_q1    = xbar_q1 + moe_q1
(lo_q1, hi_q1)
```

#### ✅ Final Conclusion — Q1

| Check | Result |
|-------|--------|
| z-stat | 2.12 |
| Critical value (±) | 1.96 |
| p-value | 0.034 |
| 95% CI | (10.02, 10.58) |

Both methods agree — the hypothesized mean of 10 mm lies **outside** the confidence interval, |z| exceeds the critical threshold, and p-value < α.

**We reject H₀.** There's sufficient statistical evidence that the true mean bolt diameter is **not 10 mm** — the factory's claim is questionable at the 5% significance level.

## 📌 Question 2 — One-Sample Proportion Z-Test

A survey of **800 voters** finds that **480 support** a new policy.
A pollster claims that **55% of the population** supports the policy.

| Parameter | Value |
|-----------|-------|
| Sample size (n) | 800 |
| Successes (x) | 480 |
| Hypothesized proportion (p₀) | 0.55 |
| Significance level (α) | 0.05 |

Perform a **two-tailed one-sample proportion Z-test**. Also compute the **95% CI**.

---

#### Step 1 — Hypothesis

- **H₀:** p = 0.55
- **H₁:** p ≠ 0.55 (two-tailed — the observed proportion could be higher or lower)

We're not trying to show the support is specifically greater or less, just whether it differs.

#### Step 2 — Why Proportion Z-Test?

- Data is count-based (successes out of n) → proportion test
- n is large → CLT applies, so we can use normal approximation
- Standard error under H₀ uses p₀ (not p̂), because we assume H₀ is true when testing

```python
# Q2 given values
x2   = 480    # number of supporters
n2   = 800    # total surveyed
p0_2 = 0.55   # claimed population proportion
a2   = 0.05   # significance level
```

```python
phat_2 = x2 / n2   # observed sample proportion
phat_2
```

#### Step 3 — Test Statistic

$$z = \frac{\hat{p} - p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}}$$

SE uses p₀ here (not p̂) because under H₀ we assume the true proportion is p₀.

```python
se_h0_2 = (p0_2 * (1 - p0_2) / n2) ** 0.5   # SE under null hypothesis
z_q2    = (phat_2 - p0_2) / se_h0_2
z_q2
```

#### Step 4 — Critical Value Method

```python
zcrit_q2 = norm.ppf(1 - a2/2)
zcrit_q2
```

**Decision:** 2.84 > 1.96 → |z| > z_critical → **Reject H₀**

#### Step 5 — p-value Method

```python
pval_q2 = 2 * norm.sf(abs(z_q2))
pval_q2
```

**Decision:** 0.0044 < 0.05 → p-value < α → **Reject H₀** ✓

#### Step 6 — 95% Confidence Interval

$$CI = \hat{p} \pm z_{\alpha/2} \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

**Note:** For CI we switch to p̂ in the SE formula — we're now estimating the true proportion, not assuming H₀.

```python
se_est_2 = (phat_2 * (1 - phat_2) / n2) ** 0.5   # SE using p̂ for CI
moe_q2   = zcrit_q2 * se_est_2
lo_q2    = phat_2 - moe_q2
hi_q2    = phat_2 + moe_q2
(lo_q2, hi_q2)
```

#### ✅ Final Conclusion — Q2

| Check | Result |
|-------|--------|
| Sample proportion (p̂) | 0.60 |
| z-stat | 2.84 |
| Critical value (±) | 1.96 |
| p-value | 0.0044 |
| 95% CI | (0.566, 0.634) |

The claimed proportion p₀ = 0.55 is **outside** the 95% CI, and p-value is well below 0.05.

**We reject H₀.** The data suggests the true support level is significantly higher than the pollster's claimed 55%.

## 📌 Question 3 — Two-Proportion Z-Test (A/B Test Style)

| Group | n | Conversions |
|-------|---|-------------|
| Control | 5000 | 600 |
| Treatment | 5000 | 680 |

Test at **α = 0.01** whether the treatment **significantly improves** conversion rate.
Also compute the **99% CI** for the difference in proportions.

**Key concept to address:** When to use **pooled** vs **unpooled** SE?

---

#### Step 1 — Hypothesis

- **H₀:** p_ctrl = p_trt (treatment has no effect)
- **H₁:** p_trt > p_ctrl (treatment improves conversion — one-tailed)

One-tailed because we specifically care whether treatment is **better**, not just different.

#### Step 2 — Pooled vs Unpooled SE — The Key Distinction

This trips people up, so worth being clear:

| Situation | SE to use | Why |
|-----------|-----------|-----|
| **Hypothesis test** | Pooled | Under H₀ we assume p_ctrl = p_trt, so we estimate a single common proportion |
| **Confidence Interval** | Unpooled | We're estimating the actual difference — H₀ is no longer assumed, each group gets its own p̂ |

```python
# Q3 given values
nc, xc = 5000, 600   # control: sample size, conversions
nt, xt = 5000, 680   # treatment: sample size, conversions
a3 = 0.01        # significance level
```

```python
# individual sample proportions
pc = xc / nc   # control proportion
pt = xt / nt   # treatment proportion
pc, pt
```

#### Step 3 — Pooled Proportion & SE (for hypothesis test)

$$\hat{p}_{pool} = \frac{x_c + x_t}{n_c + n_t}$$

```python
# pooled estimate — assumes H0: pc = pt
pp = (xc + xt) / (nc + nt)
pp
```

```python
# pooled SE for the test statistic
se_pool = (pp * (1 - pp) * (1/nc + 1/nt)) ** 0.5
se_pool
```

```python
z_q3 = (pt - pc) / se_pool
z_q3
```

#### Step 4 — Critical Value (one-tailed, α = 0.01)

One-tailed → rejection region is only on the right side.
We need z such that P(Z < z) = 0.99

```python
# one-tailed critical value at alpha=0.01
zcrit_q3_test = norm.ppf(1 - a3)
zcrit_q3_test
```

**Decision:** z = 2.39 > z_critical = 2.33 → **Reject H₀**

#### Step 5 — p-value (one-tailed)

```python
# one-tailed p-value: P(Z > z)
pval_q3 = norm.sf(z_q3)
pval_q3
```

**Decision:** p = 0.0083 < α = 0.01 → **Reject H₀** ✓

#### Step 6 — 99% Confidence Interval (Unpooled SE)

$$CI = (\hat{p}_t - \hat{p}_c) \pm z_{\alpha/2} \cdot \sqrt{\frac{\hat{p}_c(1-\hat{p}_c)}{n_c} + \frac{\hat{p}_t(1-\hat{p}_t)}{n_t}}$$

Here we use **unpooled SE** because we're estimating the actual difference between two proportions — no assumption of equality.

```python
# unpooled SE — each group uses its own proportion
se_unpool = np.sqrt(pc*(1-pc)/nc + pt*(1-pt)/nt)
se_unpool
```

```python
# two-tailed critical value for 99% CI
zcrit_q3_ci = norm.ppf(1 - a3/2)
zcrit_q3_ci
```

```python
diff_q3 = pt - pc   # observed difference in proportions
lo_q3   = diff_q3 - zcrit_q3_ci * se_unpool
hi_q3   = diff_q3 + zcrit_q3_ci * se_unpool
(lo_q3, hi_q3)
```

#### ✅ Final Conclusion — Q3

| Check | Result |
|-------|--------|
| Control p̂ | 0.12 (12%) |
| Treatment p̂ | 0.136 (13.6%) |
| z-stat (pooled SE) | 2.39 |
| Critical value (one-tailed) | 2.33 |
| p-value | 0.0083 |
| 99% CI (unpooled SE) | (-0.0012, 0.0332) |

**We reject H₀.** The treatment group shows a statistically significant improvement in conversion rate at α = 0.01.

**However**, notice that the 99% CI barely includes 0 — the lower bound is slightly negative (-0.0012). This is because the CI uses a two-tailed z (2.576) while the test was one-tailed (2.326). The two methods are answering slightly different questions here — the hypothesis test was directional, the CI is not. Both are valid, but comparing them directly can be misleading.

---
### Pooled vs Unpooled — Quick Summary

- **Pooled SE** → for the **test statistic** only, because under H₀ we assume the two groups have the *same* underlying proportion. Merging them gives a better estimate of that common p.
- **Unpooled SE** → for **confidence intervals**, because we're no longer assuming equality — we want to estimate how far apart the two proportions actually are.

## 📌 Question 4 — Two-Sample Z-Test + Power 

Drug A and Drug B are being compared for blood pressure reduction.

| Parameter | Drug A | Drug B |
|-----------|--------|--------|
| Mean reduction (μ) | 12 mmHg | 10.5 mmHg |
| Population std dev (σ) | 4 mmHg | 3.5 mmHg |
| Sample size (n) | 100 | 120 |
| Significance level (α) | 0.05 | — |


- **(a)** Test H₀: μ_A = μ_B at α = 0.05 using a two-sample Z-test
- **(b)** Compute the power of this test assuming the true difference is 2 mmHg
- **(c)** How large does n need to be (equal groups) to achieve 90% power for a 2 mmHg difference?

---

#### Step 1 — Hypotheses

- **H₀:** μ_A = μ_B (the drugs reduce BP by the same amount on average)
- **H₁:** μ_A ≠ μ_B (two-tailed — we're open to either drug being better)

Going two-tailed here because we're just testing if there's *any* difference, not claiming A is better in advance.

#### Step 2 — Why Two-Sample Z-Test?

- Both population σ values are **known** → Z-test (not t-test)
- Both sample sizes are large (100 and 120) — CLT applies comfortably
- The test statistic measures how far the observed difference is from 0, in units of standard error

```python
# given values for Q4
mu_a  = 12      # mean BP reduction for Drug A (mmHg)
mu_b  = 10.5    # mean BP reduction for Drug B (mmHg)
sig_a = 4       # population std dev, Drug A
sig_b = 3.5     # population std dev, Drug B
n_a   = 100     # sample size, Drug A
n_b   = 120     # sample size, Drug B
alpha = 0.05    # significance level
```

#### Step 3 — Standard Error and Test Statistic

$$z = \frac{(\bar{x}_A - \bar{x}_B) - 0}{\sqrt{\frac{\sigma_A^2}{n_A} + \frac{\sigma_B^2}{n_B}}}$$

The denominator (SE) tells us how much the *difference* in sample means varies just from random chance.
A large z means the observed gap is hard to explain away as noise.

```python
std_err = np.sqrt(sig_a**2 / n_a + sig_b**2 / n_b)
z_stat  = (mu_a - mu_b) / std_err

std_err, z_stat
```

#### Step 4 — Critical Value Method

Two-tailed at α = 0.05 → rejection region splits equally: 2.5% on each side.
We reject H₀ if |z_stat| > z_critical.

```python
z_crit = norm.ppf(1 - alpha / 2)
z_crit
```

**Decision:** |z| = 2.93 > z_critical = 1.96 → **Reject H₀**

#### Step 5 — p-value

Two-tailed p-value: probability of seeing a z this extreme (or more) just by chance, assuming H₀ is true.

Note: `norm.sf(abs(z))` gives P(Z > |z|), multiply by 2 for two-tailed.

```python
pval = 2 * norm.sf(abs(z_stat))
pval
```

**Decision:** p = 0.0034 < α = 0.05 → **Reject H₀** ✓

Both methods agree — the observed difference of 1.5 mmHg is statistically significant.

#### ✅ Part (a) Conclusion

| Check | Result |
|-------|--------|
| Observed difference | 1.5 mmHg |
| Standard error | 0.512 |
| z-stat | 2.93 |
| Critical value (±) | 1.96 |
| p-value | 0.0034 |

**We reject H₀.** The two drugs do not appear to reduce blood pressure by the same amount. Drug A shows a significantly greater reduction at the 5% level.

---
#### Part (b) — Computing Power

Power is the probability that our test *correctly* rejects H₀ when H₁ is actually true.

Here the question says: *assume the true difference really is 2 mmHg* — what's the chance our test picks it up?

The key idea: under H₁, the Z-statistic is no longer centered at 0. It's centered at **Δ/SE** (the "non-centrality parameter"). Power is then just the area of that shifted distribution that falls in the rejection region.

```python
true_diff = 2   # assumed true difference in means (mmHg)

# how many SEs away is the true difference from 0?
ncp = true_diff / std_err   # non-centrality parameter
ncp
```

The true distribution is centered at z ≈ 3.91 (not 0).
The cutoff is still at 1.96. So most of the true distribution is already past the cutoff — power should be high.

```python
# power = area of the TRUE distribution that lands in the rejection region
# right tail (main contributor) + left tail (negligible but included for correctness)
right_tail = norm.sf(z_crit - ncp)
left_tail  = norm.cdf(-z_crit - ncp)

power_b = right_tail + left_tail
power_b
```

#### ✅ Part (b) Conclusion

**Power ≈ 97.4%** — with n=100 and n=120, this test has a very high chance of detecting a true 2 mmHg difference.

The z_β (distance from cutoff to the true mean, in SDs) = 3.91 − 1.96 = **1.95**, which corresponds to β ≈ 2.6%. So we'd only miss the real difference about 2.6% of the time.

**-1.95** is the standardized value of the cutoff 1.96 measured from the true mean, 3.91. 

After we standardize the distribution, the values beyond -1.95 falls into rejection zone

---
#### Part (c) — Finding Sample Size for 90% Power

Now we flip the problem: instead of computing power from n, we fix the target power and solve for n.

**The key equation:**

$$\frac{\Delta}{SE} = z_{\alpha/2} + z_{\beta}$$

where z_β = norm.ppf(0.90) for 90% power. Rearranging for n (assuming equal group sizes):

$$n = \frac{(\sigma_A^2 + \sigma_B^2)(z_{\alpha/2} + z_{\beta})^2}{\Delta^2}$$

Why do we *add* z_α/2 and z_β? They measure the distance from the null mean to the cutoff, and from the cutoff to the true mean respectively. Together they span the full gap the two distributions need to be apart.

```python
beta   = 0.10   # 1 - 0.90 power target

z_a = norm.ppf(1 - alpha / 2)   # 1.96
z_b = norm.ppf(1 - beta)        # equal to norm.ppf(power=0.90)

z_a, z_b
```

```python
n_req = (z_a + z_b)**2 * (sig_a**2 + sig_b**2) / true_diff**2
n_req
```

```python
n_final = math.ceil(n_req)
n_final
```

#### ✅ Part (c) Conclusion

| Component | Value |
|-----------|-------|
| z_α/2 (two-tailed, α=0.05) | 1.96 |
| z_β (for 90% power) | 1.28 |
| z_α/2 + z_β | 3.24 |
| σ_A² + σ_B² | 28.25 |
| Δ (to detect) | 2 mmHg |
| **n per group** | **75** |

We need **n = 75 per group** — considerably less than the 100 and 120 in the original study. The original setup was actually over-powered for this specific effect size, which is why part (b) gave us 97.4% — well above the 90% target.
