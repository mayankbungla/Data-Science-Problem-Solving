```python
import numpy as np
from scipy.stats import norm
```

---
## 📌 Question 1

A factory claims that the mean diameter of its bolts is **10 mm**.

A sample of **50 bolts** is taken, and the following statistics are observed:

- Sample mean (**x̄**) = **10.3 mm**
- Population standard deviation (**σ**) = **1.0 mm**
- Significance level (**α**) = **0.05**

Perform a **two-tailed one-sample Z-test** to evaluate the factory’s claim.

- Compute the **Z-statistic**
- Compare it with the **critical value**
- Clearly state your **conclusion**
- Also compute the **95% confidence interval** for the mean
---

#### Step 1 : Hypothesis
- H₀: μ = 10
- H₁: μ ≠ 10 (two-tailed test)

####  Step 2: Approach

- Population σ is known → use Z-test  
- Two-tailed test → check both directions  
- Compute:
  - Test statistic (z)
  - Critical value
  - p-value

```python
# Given values
x_bar = 10.3
mu = 10
sigma = 1
n = 50
alpha = 0.05
```

#### Step 3: Compute Test Statistic

Formula: z = (x̄ - μ₀) / (σ / √n)

```python
z = (x_bar - mu) / (sigma/ n**0.5)
z
```

#### Step 4: Critical Value Method

For α = 0.05 (two-tailed):

α/2 = 0.025  
Left area = 1 - 0.025 = 0.975  

We need:
z such that P(Z < z) = 0.975

```python
# Critical z-value
z_critical = norm.ppf(1-alpha/2)
z_critical
```

##### Decision
- |z| > z_critical 
- Reject H₀

#### Step 5: p-value Method

The p-value is computed as:

p = 2 × P(Z > |z|)

We use the survival function (sf), which directly gives the right-tail probability.

```python
# p-value (two-tailed)
p = 2 * norm.sf(abs(z))
p
```

##### Decision
- p < alpha
- Reject H₀

#### Step 6: Confidence Interval

CI = x̄ ± z_critical × (σ / √n)

```python
margin = z_critical * (sigma / np.sqrt(n))
lower = x_bar - margin
upper = x_bar + margin

(lower, upper)
```

## Final Conclusion

- Test statistic: z ≈ 2.12
- Critical value: ±1.96
- p-value ≈ 0.034

Since:
- |z| > z_critical
- p < α
- 10 is outside the confidence interval

We reject H₀.

The data provides sufficient evidence that the true mean diameter differs from 10 mm.
