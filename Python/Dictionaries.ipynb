```python
 # Given a string, build a dictionary mapping each character to its count.
s = 'hello'
from collections import Counter
s = Counter(s)
s
```

```python
s = 'banana'
d = dict()
for i in s:
    if i in d:
        d[i] +=1
    else:
        d[i] = 1
print(d)
```

```python
# Problem. Count the frequency of each word in a paragraph, case-insensitive, ignoring punctuation.
import string
x = 'The cat. The dog! The cat?'
for p in string.punctuation:
    x = x.replace(p,'')
x = Counter(x.split())
x
```

```python
# Problem. Given a list of (student, subject, score) tuples, build a nested dictionary {student: {subject: score}}.
x = [('Alice','Math',90),('Bob','Math',80),('Alice','Sci',85)]
d = dict()

for a,b,c in x:
    if a not in d:
        d[a] = {}
    d[a][b] = c
d
```

```python
# Sort a dictionary by its values in descending order. Return a list of (key, value) tuples.
x= {'a':3,'b':1,'c':2}
result = sorted(x.items(), key = lambda x:x[1], reverse = True)
result
```
