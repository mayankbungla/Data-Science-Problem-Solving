```python
# Problem. Given a list of (student, subject, score) tuples, build a nested dictionary {student: {subject: score}}.
x = [('Alice','Math',90),('Bob','Math',80),('Alice','Sci',85)]
d = dict()

for a,b,c in x:
    if a not in d:
        d[a] = {}
    d[a][b] = c
d
```

```python
# Given two dictionaries representing inventory before and after, return a dictionary of items whose quantity changed and by how much.
x = {'a':5, 'b':3, 'c':10, 'e':7}
y = {'a':5, 'b':4, 'c':8, 'd':2}

changes = {}
for key in x|y:      # Union of keys
    if y.get(key,0) != x.get(key,0):                # To ignore those without any changes
        changes[key] = y.get(key,0) - x.get(key,0)

changes
```

```python
 # Group a list of strings by their first character into a dictionary.
lst = ['apple','ant','bat','banana','cat']
d = {}
for x in lst:
    if x[0] not in d:
        d[x[0]] = [x]
d[x[0]].append(x)
d
```

```python
from collections import defaultdict
d = defaultdict(list)           # creates empty list for missing key

for x in lst:
    d[x[0]].append(x)
d
```

```python
# Given a deeply nested dictionary, write a function that flattens it. Keys in the result are paths joined by dots.
x =  {'a':1,'b':{'c':2,'d':{'e':3}}}
def flatten(x, pre = ''):
    res = {}
    for k,v in x.items():
        key = f'{pre}.{k}' if pre else k
        if isinstance(v,dict):
            res.update(flatten(v,key))
        else:
            res[key] = v
    return res
flatten(x)
```

```python
# Given a list of dicts representing transactions [{'user':'A','amount':100}, ...], compute the total amount per user.
x = [{'user':'A','amount':100},{'user':'B','amount':50},{'user':'A','amount':25}]

res = defaultdict(int)
for i in x:
    res[i['user']] += i['amount']
res
```
