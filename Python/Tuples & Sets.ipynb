```python
# Problem. Given two lists representing inventory before and after a sale, return three sets: items sold out,
# items still available, and newly stocked items.

before=['a','b','c']
after=['b','d']

def func(b,a):
    b_set = set(b)
    a_set = set(a)
    
    sold = b_set - a_set
    still_available = b_set & a_set
    new = a_set - b_set
    return sold, still_available, new
func(before,after)
```

```python
# You have a tuple of immutable records (id, score). Given a list of such tuples, return the record with the highest score. 
# Break ties by the smallest id.

arr =  [(1,80),(2,90),(3,90)]
best = arr[0]
for i in  arr:
    if i[1]>best[1]:
        best = i
    elif i[1] == best[1] and i[0] < best[0]:
        best = i
best
```
