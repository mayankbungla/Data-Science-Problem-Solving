```python
# Given a list of tuples (name, age), find the average age of all people whose name starts with a vowel.
lst =  [('Alice',30),('Bob',25),('Eve',35)]

arr = [age for name,age in lst if name[0].upper() in 'AEIOU']
sum(arr)/len(arr)
```
