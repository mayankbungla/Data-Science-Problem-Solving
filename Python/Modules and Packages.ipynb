```python
 # Use the 'math' module to compute the square root of 2, the value of pi, and sin(pi/2).
import math

print(math.sqrt(2))
print(math.pi)
print(math.sin(math.pi/2))
```

```python
 # Use the 'random' module to pick 5 unique random numbers between 1 and 100.
import random
random.sample(range(1,101),5)
```

```python
# Print the current date and time formatted as 'YYYY-MM-DD HH:MM:SS'
from datetime import datetime
datetime.now().strftime("%Y-%m-%d %H:%M:%S")
```

```python
# Use 'os' to list all files in the current directory that end with '.ipynb'.
import os
for file in os.listdir():
    if file.endswith('.ipynb'):
        print(file)
```

```python
# Use 'pathlib' to count how many '.ipynb' files exist anywhere under a given directory.
from pathlib import Path
directory = Path('.')
print(sum(1 for file in directory.rglob('.ipynb')))
```
