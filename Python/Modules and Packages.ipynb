```python
 # Use 'functools.lru_cache' to memoize a recursive function (e.g., Fibonacci) and compare runtime to the uncached version for n=35
from functools import lru_cache
import time

# without Cache

def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)

# with Cache
@lru_cache(maxsize = None)

def fib_cached(n):
    if n<=1:
        return n
    return fib_cached(n-1) + fib_cached(n-2)
    
n = 35
# Time the uncached version
start = time.perf_counter()
result1 = fib(n)
end = time.perf_counter()
print(f"Without cache: Fibonacci({n}) = {result1}")
print(f"Time taken: {end - start:.6f} seconds")


# Time the cached version
start = time.perf_counter()
result2 = fib_cached(n)
end = time.perf_counter()

print(f"\nWith cache: Fibonacci({n}) = {result2}")
print(f"Time taken: {end - start:.6f} seconds")
```
