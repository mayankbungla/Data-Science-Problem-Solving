```python
# Problem. Read three integer scores and print their average to 2 decimal places.
a = int(input("Enter number a :"))
b = int(input("Enter number b :"))
c = int(input("Enter number c :"))
average = (a+b+c)/3
print(f'{average:.2f}')
```

```python
# Problem. Read two values from a single line of input separated by a space; one is an int, one is a float.
# Detect and store each in the correct type without knowing the order.

a,b = input('Enter integer and float number').split(",")

try:
    a= int(a)
    b= float(b)
    print(f'float={b}, int={a}')
except ValueError:
    a= float(a)
    b= int(b)
    print(f'float={a}, int={b}')
```
