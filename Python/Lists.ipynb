```python
# Problem. Flatten a one-level nested list of lists into a single list.
lst= [[1,2],[3,4],[5]]

def flatten(lst):
    result = []
    for i in lst:
        if isinstance(i,list):
            result.extend(flatten(i))
        else:
            result.append(i)
    return result

flatten(lst)
```

```python
# Problem. Remove duplicates from a list while preserving the order of first occurrence.
lst =  [3, 1, 3, 2, 1, 4]
result = []
seen = set()
for i in lst:
    if i not in seen:
        result.append(i)
        seen.add(i)
result
```

```python
# Problem. Given a list of numbers, return a new list where each element is replaced by the running cumulative sum up to that index.
lst =  [1, 2, 3, 4]
result = []
total =0
for i in lst:
    total +=i
    result.append(total)
result
```

```python
# Problem. Find the second largest unique element in a list.
lst = [4, 1, 4, 3, 2]

def func(lst):
    unique = sorted(set(lst),reverse = True)
    if len(unique) <2:
        return "Not enough unique numbers"
    return unique[1]
func(lst)
```

```python
#  Given a list of integers, return all pairs (a, b) with a < b such that a + b == target. Each input integer may appear in 
# multiple pairs only if there are duplicates.
lst = [1, 3, 2, 4, 3]
target=5

def func(lst, target):
    result = []
    for i in range(len(lst)):
        for j in range(i+1,len(lst)):
            if lst[i] + lst[j] == target:
                a,b = sorted((lst[i], lst[j]))
                result.append((a,b))
    return result
func(lst, target)    
```

```python
# Problem. Given a list, return the list of elements that appear more than once, sorted in descending order of their first occurrence.
from collections import Counter

arr = [4, 5, 4, 6, 7, 5, 8, 7]
count= Counter(arr)
new = [x for x in count if count[x]>1]
new.sort(key = lambda x : arr.index(x), reverse = True)
new
```
