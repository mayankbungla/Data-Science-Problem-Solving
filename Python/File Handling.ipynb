```python
#  Read a text file and print the number of lines, words, and characters (like Unix 'wc')

with open("C://Users//JackSparrow//Documents//greeting.txt", "r") as f:
    lines = 0
    words = 0
    chars = 0
    for i in f:
        lines += 1
        words += len(i.split())
        chars += len(i)
    print(f'lines = {lines}, words = {words}, characters = {chars}')
    
```

```python
# Read a CSV file with columns name,age,city and print each row as a dict
import csv

with open("C://Users//JackSparrow//Documents//people.csv") as f:
    reader = csv.DictReader(f)

    for row in reader:
        print(row)
```

```python
# Given a list of dictionaries with the same keys, write them out to a CSV file with the keys as header.
lst = [{'a':1,'b':2},{'a':3,'b':4}] 
with open("C://Users//JackSparrow//Documents//output.csv", "w", newline= "") as f:
    writer = csv.DictWriter(f, fieldnames = lst[0].keys())

    writer.writeheader()
    writer.writerows(lst)
```
