```python
# . Given two CSVs 'orders.csv' and 'customers.csv', join them on customer_id and write the result to 'joined.csv'. No pandas allowed.

#         Customer
customers = {}
with open("C://Users//JackSparrow//Documents//customers.csv", "r") as c:
    reader = csv.DictReader(c)
    for row in reader:
        customers[row['customer_id']] = row

#         Order
orders= []
with open("C://Users//JackSparrow//Documents//orders.csv", "r") as o:
    reader = csv.DictReader(o)
    for row in reader:
        orders.append(row)

#         Joined
with open("C://Users//JackSparrow//Documents//joined.csv", "w", newline= "") as j:
    fieldnames = ["order_id", "customer_id", "name", "amount"]
    writer = csv.DictWriter(j, fieldnames=fieldnames)
    writer.writeheader()

    for order in orders:
        cid = order["customer_id"]
        customer = customers[cid]

        joined_row = {
        "order_id": order["order_id"],
        "customer_id":cid,
        "name": customer["name"],
        "amount": order["amount"]
        }
        writer.writerow(joined_row)
        
with open("C://Users//JackSparrow//Documents//joined.csv", "r") as f:
    for line in f:
        print(line)
```

```python
# Read a configuration file in 'key=value' format (ignore blank lines and lines starting with #). Return a dictionary of the settings.

settings = {}
with open("C://Users//JackSparrow//Documents//config.txt", "r") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue

        if line.startswith('#'):
            continue

        key,val = line.split("=")
        settings[key] = val
settings    
```

```python
# Given a large text file, write a function that returns the n most frequent words (lowercased) without
# loading the whole file into memory at once

from collections import Counter

def top_n_words(filepath,n):
    counts = Counter()
    with open(filepath, 'r') as f:
        for line in f:
            counts.update(line.lower().split())
    return counts.most_common(n)

top_n_words("C://Users//JackSparrow//Documents//words.txt",10)
```
