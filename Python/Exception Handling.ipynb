```python
# Define a custom exception 'InvalidAgeError'. Raise it when a function receives age < 0 or > 150
class InvalidAgeError(Exception):
    pass

def validate_age(age):
    if age < 0 or age > 150:
        raise InvalidAgeError("age must be between 0 and 150.")
    print("Valid Age")
    
try:
    age= int(input("Enter Age : "))
    validate_age(age)
except InvalidAgeError as e:
    print(e)
except ValueError:
    print('Enter a Numeric Value')
```

```python
# catch a low-level library exception and replace it with a meaningful application-specific exception that preserves the original cause.

class UserDataError(Exception):
    pass

def validate_age(str_age):
    try:
        age = int(str_age)
        return age
    except ValueError as e:
        raise UserDataError("Invalid Age Entered") from e

try:
    age = validate_age('abc')
    print(age)
except UserDataError as e:
    print(e)
```

```python
# Implement a retry decorator that retries a function up to N times if it raises any exception, with a short delay between attempts.
import time

def retry(max_attempts,delay):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for attempt in range(1,max_attempts+1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f'Attempt {attempt} failed : {e}')
                    if attempt == max_attempts:
                        raise 
                    time.sleep(delay)
        return wrapper
    return decorator

@retry(3,1)

def divide():
    return 10/0
divide()
```
