```python
# Write a function that takes any number of positional arguments and returns their sum.

def sum_all(*args):
    return sum(args)
sum_all(1,2,3,4)
```

```python
# Write a function that accepts any number of keyword arguments and returns a dictionary excluding any keys whose value is None.
def func(**kwargs):
    return {k:v for k,v in kwargs.items() if v !=None}
         
func(a=1, b=None, c=3)
```

```python
# Problem. Write a recursive function to compute the factorial of n. (Basic recursion only.)

def factorial(n):
    if n==0 or n==1:
        return 1
    return n* factorial(n-1)
factorial(6)    
```

```python
# Write a function 'make_counter' that returns a new function. Each call to the returned function increments and returns a counter starting from 0.
def make_counter():
    count = 0
    
    def counter():
        nonlocal count 
        count +=1
        return count
    return counter
    
c = make_counter()
print(c())
print(c())
print(c())
```

```python
# Demonstrate the mutable default argument bug. Write a function 'append_to(x, lst=[])' that should append x to a fresh list each time but doesn't. 
# Then fix it.


# BUGGY VERSION

def append_to(x, lst=[]):
    lst.append(x)
    return lst
    
append_to(1) 
append_to(2) 
append_to(3) 
```

```python
# FIXED

def func(n, lst = None):
    if lst is None:
        lst = []
    lst.append(n)
    return lst
    
print(func(1))
print(func(2))
print(func(3))
```
