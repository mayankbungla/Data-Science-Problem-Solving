```python
# Problem. Write a recursive function to compute the factorial of n. (Basic recursion only.)

def factorial(n):
    if n==0 or n==1:
        return 1
    return n* factorial(n-1)
factorial(6)    
```

```python
# Write a function 'make_counter' that returns a new function. Each call to the returned function increments and returns a counter starting from 0.
def make_counter():
    count = 0
    
    def counter():
        nonlocal count 
        count +=1
        return count
    return counter
    
c = make_counter()
print(c())
print(c())
print(c())
```

```python
# Write a function that takes another function and a list, and returns a list of the function's results applied to each element. Don't use map

def apply(func,lst):
    res = []
    for i in lst:
        res.append(func(i))
    return res
apply(lambda x: x*2 , [1,2,3])
```

```python
# Write a decorator 'timer' that prints how long the decorated function takes to run.
import time

def timer(func):
    def wrapper(*args, **kwargs):
        
        start = time.time()
        res = func(*args, **kwargs)
        end = time.time()

        print(f'{func.__name__} took {start - end:.4f} seconds to execute')
        return res
    return wrapper

@timer

def slow_func():
    time.sleep(2)
    print('Done')
slow_func()
```

```python
# Write a function that takes a list of functions and a value, then returns the result of applying them sequentially (function composition)

def compose(lst,val):
    for func in lst:
        val = func(val)
    return val
compose([str, lambda s: s+'a!', str.upper], 4)
```

```python
# Demonstrate variable scope: write a function that has a local variable 'x', a nested function that reads it (closure), and another nested 
# function that needs to modify it. Show the difference between nonlocal and global.
x = 10
def outer():
    x = 20
    def inner():
        print(f'inner x initial : {x}')
    def modify_nonlocal():
        nonlocal x
        x+=5
        print(f'inner x modified : {x}')
    def modify_global():
        global x
        x -= 3
        print(f'Global x Modified : {x}')

    inner()
    modify_nonlocal()
    modify_global()

outer()
print(f'Global x after outer : {x}')
```
