```python
# Write a generator function 'count_down(n)' that yields numbers from n down to 1

def count_down(n):
    while n>0:
        yield n
        n -= 1
for i in count_down(5):
    print(i)
```

```python
# Write a generator that yields the Fibonacci sequence indefinitely

def fibonacci():
    a = 0
    b = 1
    while True:
        yield a
        a,b = b, a+b

fib =  fibonacci()
for _ in range(10):
    print(next(fib))
```

```python
# Implement an iterator class 'EvenNumbers' that yields even numbers up to a given limit.

class EvenNumbers:

    def __init__(self, limit):
        self.limit = limit
        self.current = 2

    def __iter__(self):
        return self

    def __next__(self):
        if self.current > self.limit:
            raise StopIteration
        even = self.current
        self.current +=2
        return even
        
evens = EvenNumbers(10)

for num in evens:
    print(num)
```

```python
# Use itertools.groupby to group a sorted list of (category, item) tuples by category.
from itertools import groupby

data = [
    ("Fruit", "Apple"),
    ("Fruit", "Banana"),
    ("Fruit", "Orange"),
    ("Vegetable", "Carrot"),
    ("Vegetable", "Potato"),
    ("Vegetable", "Tomato")
]

for category, items in groupby(data, key = lambda x: x[0]):
    print(category)

    for item in items:
        print(item)
    print()
```

```python
# Implement a generator 'windowed(iterable, n)' that yields successive tuples of length n (sliding window).
def windowed(iterable, n):
    for i in range(len(iterable) - n +1):
        yield tuple(iterable[i:i+n])

numbers = [1, 2, 3, 4, 5]

for window in windowed(numbers, 3):
    print(window)
```

```python
from collections import deque
# Works on generator objects too
def windowed(iterable,n):
    it = iter(iterable)
    window = deque(maxlen = n)

    for _ in range(n):
        try:
            window.append(next(it))
        except StopIteration:
            return
    yield tuple(window)
    for item in it:
        window.append(item)
        yield tuple(window)

numbers = [1, 2, 3, 4, 5]

for window in windowed(numbers, 3):
    print(window)
```

```python
# flatten an arbitrarily nested list of integers.

def func(lst):
    res = []
    for i in lst:
        if isinstance(i,list):
            res.extend(func(i))
        else:
            res.append(i)
    return res
x=  [1,[2,[3,4],5],6]
func(x)
```

```python
# Use 'yield from' to flatten an arbitrarily nested list of integers.

def flatten(lst):
    for item in lst:
        if isinstance(item,list):
            yield from flatten(item)
        else:
            yield item
nested = [1, [2, 3], [4, [5, 6]], 7]

for num in flatten(nested):
    print(num)
```
