```python
# Problem. Count the number of vowels (a, e, i, o, u) in a string, case-insensitive.

def func(sent):
    count = 0
    vovel = 'aeiouAEIOU'
    for char in sent:
        if char in vovel:
            count +=1
    return count
func("Beautiful")
```

```python
# Problem. Check if a string is a palindrome (reads the same forwards and backwards), ignoring case and non-alphanumeric characters.
def func(sent):
    cleaned = ''.join(x.lower() for x in sent if x.isalnum())
    if cleaned == cleaned[::-1]:
        return 'True'
    else:
        return 'False'
a = 'A man, a plan, a canal: Panama'

print(func(a))
```

```python
# Given a string, return the most frequent character. If there's a tie, return the one that appears first.
a = 'Horror Movie'

def func(sent):
    d = {}
    for i in sent:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    max_count = max(d.values())
    for i,j in d.items():
        if d[i] == max_count:
            return i,j

    
print(func(a))
```

```python
# Problem. Check whether two strings are anagrams of each other (same letters, possibly different order)
from collections import Counter
a = 'listen'
b = 'Silent'

def func(a,b):
    # return sorted(a) == sorted(b)
    return Counter(a.lower()) == Counter(b.lower())

print(func(a,b))
```

```python
# Given a string, find the length of the longest substring without repeating characters. 
sent =  'abcabcbb'

def func(sent):
    max_len = 0
    longest_substring = ''

    for i in range(len(sent)):
        seen = set()
        for j in range(i,len(sent)):
            if sent[j] in seen:
                break
            seen.add(sent[j])
            curr_count = j-i +1
            if curr_count > max_len:
                max_len = curr_count
                longest_substring = sent[i:j+1]
    return max_len, longest_substring

print(func(sent))
```

```python
# 20. Remove all duplicate characters from a string while preserving the order of first occurrence.
sent = 'programming'

def func(sent):
    seen = set()   # For more efficiency using set ... time complexity -- O(n)
    result = ''
    for i in sent:
        if i not in seen:
            seen.add(i)
            result += i
    return result

print(func(sent))
```

```python
#  21.  Given a string of words, return a dictionary mapping each word length to a list of words of that length, 
# with words alphabetically sorted within each group

sent = 'the quick brown fox jumps'

def func(sent):
    d = dict()
    cleaned = sent.split(' ')
    for i in cleaned:
        l = len(i)
        if l not in d:
            d[l] = []
        d[l].append(i)
    for k in d:
        d[k].sort()
    return d

print(func(sent))
```

```python
# 22. Write a function that takes a string and returns the first non-repeating character. If none exists, return None.
sent = 'aabbccdee'

def func(sent):
    seen = {}
    for i in sent:
        if i not in seen:
            seen[i] = 1
        else:
            seen[i] += 1
    for j in seen:
        if seen[j] == 1:
            return j
    return None

print(func(sent))
```

```python
# 23.  Validate whether a string is a properly formed numeric value (integer, float, or scientific notation), without using try/except.
import re
x = '-1.23e5'
y = '1e2.3'

def func(sent):
    pattern = r'^[+-]?(\d+(\.\d*)?|\.\d+)([eE][+-]?\d+)?$'
    return bool(re.fullmatch(pattern,sent))

print(func(x))
print(func(y))
```
