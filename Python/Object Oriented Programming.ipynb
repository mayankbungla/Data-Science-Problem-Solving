```python
 # Define a class 'Rectangle' with width and height. Add a method 'area()' that returns width * height
class Rectangle():
    def __init__(self,b,h):
        self.b = b
        self.h = h
    def area(self):
        return f'Area = {self.b * self.h}'

rect = Rectangle(4,3)
print(rect.area())
```

```python
# Add a __str__ method to the Rectangle class that returns 'Rectangle(width, height)'.
class Rectangle():
    def __init__(self,b,h):
        self.b = b
        self.h = h
    def area(self):
        return f'Area = {self.b * self.h}'
    def __str__(self):
        return f'Rectangle({self.b}, {self.h})'

rect = Rectangle(4,3)
print(rect)
```

```python
# Define a class 'BankAccount' with balance. Add deposit and withdraw methods. Withdraw should raise an error if it would result in negative balance.

class BankAccount:
    def __init__(self,balance = 0):
        self.balance = balance
    def deposit(self, amount):
        if amount >0:
            self.balance += amount
    def withdraw(self, amount):
        if amount>self.balance:
            return 'Insufficient Balance'
        self.balance -= amount
        
acc = BankAccount(1000)
acc.deposit(500)
print(acc.balance)      

acc.withdraw(700)
print(acc.balance)      

acc.withdraw(1000)
```

```python
# Make 'balance' in BankAccount a read-only property. The class should only allow modifying it via deposit/withdraw
class BankAccount:
    def __init__(self,balance = 0):
        self._balance = balance

    @property
    def balance(self):
        return self._balance
        
    def deposit(self, amount):
        if amount >0:
            self._balance += amount
    def withdraw(self, amount):
        if amount>self._balance:
            return 'Insufficient Balance'
        self._balance -= amount

acc = BankAccount(1000)

print(acc.balance)   

acc.deposit(500)
print(acc.balance)   

acc.withdraw(300)
print(acc.balance)   

acc.balance = 2000  
```

```python
# Problem. Override __eq__ and __hash__ on a Point class so two Points are equal when their x and y are equal.

class Point:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def __eq__(self, other):
        if not isinstance(other, Point):
            return "Not a Point Class"
        return self.a == other.a and self.b == other.b
    def __hash__(self):
        return hash((self.a,self.b))

p1= Point(2,3)
p2= Point(2,3)
p1==p2
```

```python
# Define a class 'Counter' with __call__ so that instances can be called like functions to increment and return a count.

class Counter:
    def __init__(self):
        self.count = 0

    def __call__(self):
        self.count +=1
        return self.count
c = Counter()
print(c())   
print(c())   
print(c())   
```

```python
# Build an abstract base class 'Shape' with abstract method 'area'. Implement Circle and Triangle subclasses

from abc import ABC,abstractmethod
class Shape(ABC):     # ABC defines base class (Blueprint)
    @abstractmethod   # Every subclass must implement this function (own individual)
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius**2

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side**2

c = Circle(5)
s = Square(6)

print(c.area())
print(s.area())
```

```python
 # Implement __add__ on a Vector class so two Vectors can be summed with '+'. Vectors are 2D (x,y)

class Vector:
    def __init__(self, x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        if not isinstance(other,Vector):
            return "Not a Vector"
        return Vector(self.x + other.x, self.y + other.y)
    def __repr__(self):
        return f"Vector({self.x},{self.y})"
        
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2
print(v3)    
```

```python
# Build an 'Employee' class hierarchy: base Employee, then SalariedEmployee and HourlyEmployee subclasses. Each has a different calculate_pay() method. 
# Demonstrate polymorphism by computing pay for a list of mixed employees.

class Employee(ABC):
    def __init__(self, name):
        self.name = name
    @abstractmethod
    def calculate_pay(self):
        pass

class SalariedEmployee(Employee):
    def __init__(self,name, monthly_sal):
        super().__init__(name)
        self.monthly_sal = monthly_sal

    def calculate_pay(self):
        return self.monthly_sal

class HourlyEmployee(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        super().__init__(name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked
        
    def calculate_pay(self):
        return self.hourly_rate * self.hours_worked

employees = [
    SalariedEmployee("Alice", 5000),
    HourlyEmployee("Bob", 20, 160),
    SalariedEmployee("Charlie", 6000),
    HourlyEmployee("David", 25, 120)
]


for employee in employees:
    print(f"{employee.name}: ${employee.calculate_pay()}")
```
