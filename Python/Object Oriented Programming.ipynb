```python
 # Define a class 'Rectangle' with width and height. Add a method 'area()' that returns width * height
class Rectangle():
    def __init__(self,b,h):
        self.b = b
        self.h = h
    def area(self):
        return f'Area = {self.b * self.h}'

rect = Rectangle(4,3)
print(rect.area())
```

```python
# Add a __str__ method to the Rectangle class that returns 'Rectangle(width, height)'.
class Rectangle():
    def __init__(self,b,h):
        self.b = b
        self.h = h
    def area(self):
        return f'Area = {self.b * self.h}'
    def __str__(self):
        return f'Rectangle({self.b}, {self.h})'

rect = Rectangle(4,3)
print(rect)
```

```python
# Create a 'Square' class that inherits from Rectangle. Constructor takes only one argument 'side'.
class Square(Rectangle):
    def __init__(self,s):
        super().__init__(s,s)
    
a = Square(3)
a.area()
```

```python
# Define a class 'BankAccount' with balance. Add deposit and withdraw methods. Withdraw should raise an error if it would result in negative balance.

class BankAccount:
    def __init__(self,balance = 0):
        self.balance = balance
    def deposit(self, amount):
        if amount >0:
            self.balance += amount
    def withdraw(self, amount):
        if amount>self.balance:
            return 'Insufficient Balance'
        self.balance -= amount
        
acc = BankAccount(1000)
acc.deposit(500)
print(acc.balance)      

acc.withdraw(700)
print(acc.balance)      

acc.withdraw(1000)
```

```python
# Make 'balance' in BankAccount a read-only property. The class should only allow modifying it via deposit/withdraw
class BankAccount:
    def __init__(self,balance = 0):
        self._balance = balance

    @property
    def balance(self):
        return self._balance
        
    def deposit(self, amount):
        if amount >0:
            self._balance += amount
    def withdraw(self, amount):
        if amount>self._balance:
            return 'Insufficient Balance'
        self._balance -= amount

acc = BankAccount(1000)

print(acc.balance)   

acc.deposit(500)
print(acc.balance)   

acc.withdraw(300)
print(acc.balance)   

acc.balance = 2000  
```
